```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("test-results/compilation.csv")
plt.figure(figsize=(16,5))
ax = sns.barplot(data=df,orient="h", width=0.1,)
ax.set(xlabel='Time to Compile (in seconds, LIB)')
# plt.show()
fig = ax.get_figure()
fig.savefig("compilation.png") 
```

```python
df=pd.read_csv("test-results/lz4-compression.csv")
plt.figure(figsize=(5,2))
ax = sns.barplot(data=df,orient="h", width=0.1,)
ax.set(xlabel='Compression speed (in MB/s, HIB)')
# plt.show()
fig = ax.get_figure()
fig.savefig("lz4-compression.png") 
```

```python
df=pd.read_csv("test-results/lz4-decompression.csv")
plt.figure(figsize=(5,2))
ax = sns.barplot(data=df,orient="h", width=0.1,)
ax.set(xlabel='Decompression speed (in MB/s, HIB)')
# plt.show()
fig = ax.get_figure()
fig.savefig("lz4-decompression.png") 
```
