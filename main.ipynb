```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(rc={'figure.figsize':(4,2)})
def plot(filename,title):
    df=pd.read_csv(filename)
    # consider adding the hue back
    avg_df = df.groupby('benchmark_name')['benchmark_result'].mean().reset_index()
    g1=sns.barplot(data=avg_df,y='benchmark_name', x='benchmark_result',orient="horizontal")
    g1.set(xlabel=None)
    g1.set(ylabel=None)
    plt.legend([],[], frameon=False)
    plt.margins(y=1)
    plt.title(title)
    # plt.legend(prop={"size":6},bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
    plt.show()
```

```python
plot("memory.csv","raas/mbw MEMCPY 1024 bytes, MiB/s, HIB")
plot("io.csv","sqlite 2500 insertions timing, seconds, LIB")
plot("compilation.csv","linux kernel compilation (defconfig), seconds, LIB")
```
